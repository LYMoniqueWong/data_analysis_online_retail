##### 2. Top 5 Countries by Sales

```python
top_5_countries = df4.groupby('Country')['TotalSales'].sum().sort_values(ascending=False).head(5)
top_5_countries
```

```python
top_5_countries.plot(kind='barh', title='Top 5 Countries by Total Sales')
plt.xlabel('Total Sales')
plt.ylabel('Country')
plt.grid(axis='x')
plt.show()
```

**Insights**

1. UK has the highest sales (around 9 million)

2. Netherlands, EIRE, Germany and France are the next 4 countries each having a sales of more than 2 million

Since these countries cover the major sales revenues, the company should pay special attention to customers in these countries and make sure their product quality and service are the best. Also to break dependancy of sales from a single country they can focus on expanding sales in other countries as well.

```python
country_sales = df4.groupby('Country')['TotalSales'].sum()
total_sales = country_sales.sum()

top_5_countries = country_sales.sort_values(ascending=False).head(5)
percentages = (top_5_countries / total_sales) * 100

plt.figure(figsize=(10, 6))
bars = plt.barh(top_5_countries.index, percentages)
plt.xlabel('Percentage Contribution to Total Sales')
plt.ylabel('Country')
plt.title('Top 5 Countries by Percentage Contribution to Total Sales')
plt.grid(axis='x')
plt.xticks(rotation=45)

# Add percentage labels
for bar, percentage in zip(bars, percentages):
  plt.text(bar.get_width() + 0.5, bar.get_y() + bar.get_height() / 2, f'{percentage:.1f}%', va='center')

plt.show()
```

##### 4. RFM Analysis (Recency, Frequency, Monetary)

```python
df4['InvoiceDate'].max()
```

```python
current_dt = df4['InvoiceDate'].max() + pd.Timedelta(days=1)
current_dt
```

```python
rfm = df4.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (current_dt - x.max()).days,
    'InvoiceNo': 'count',
    'TotalSales': 'sum'
})

rfm.columns = ['Recency', 'Frequency', 'Monetary']
rfm.head(5)
```

```python
df4[df4.CustomerID==12347]['TotalSales'].sum()
```

```python
# Segment customers based on RFM
rfm['R_Segment'] = pd.qcut(rfm['Recency'], 4, labels=[4, 3, 2, 1]) # most recent customer gets higher score
rfm['F_Segment'] = pd.qcut(rfm['Frequency'], 4, labels=[1, 2, 3, 4])
rfm['M_Segment'] = pd.qcut(rfm['Monetary'], 4, labels=[1, 2, 3, 4])
rfm['RFM_Score'] = rfm[['R_Segment', 'F_Segment', 'M_Segment']].sum(axis=1)
rfm
```

```python
# Customer with highest RFM Scores
rfm.sort_values('RFM_Score', ascending=False)
```

##### 5. Customer Churn Analysis

```python
df4.head(2)
```

```python
# Create a basket matrix for association rule mining
customer_last_purchase = df4.groupby('CustomerID')['InvoiceDate'].max()
customer_last_purchase.head(5)
```

```python
type(customer_last_purchase)
```

```python
current_dt
```

```python
customer_last_purchase = (current_dt - customer_last_purchase).dt.days
customer_last_purchase.head(5)
```

```python
# Define churn threshold (e.g. 90 days w/o purchase)
churn_threshold = 90
churned_customer = customer_last_purchase[customer_last_purchase > churn_threshold]
churned_customer.head(5)
```

```python
print("Number of churned customers: ", len(churned_customer))
plt.figure(figsize=(10, 6))
plt.hist(customer_last_purchase, bins=50, color='red', alpha=0.7)
plt.axvline(x=churn_threshold, color='black', linestyle='--', linewidth=2)
plt.title('Customer Churn Distribution')
plt.xlabel('Days Since Last Purchase')
plt.ylabel('Number of Customers')
plt.show()
```
